```python
gender_wise_placement=df.groupby(["MBA_specialisation","status","gender"]).size().reset_index(name="count")
pivot_data=gender_wise_placement.pivot(index="status",columns=["MBA_specialisation","gender"],values="count")

axis=pivot_data.plot(kind="bar",edgecolor="black")
for container in axis.containers:
    axis.bar_label(container,label_type="edge")

plt.xlabel(["MBA_specialisation","gender"])
plt.ylabel("count")
plt.xticks(rotation=0)
plt.title("Gender wise placement distribution based on MBA specilization")
plt.show()
```
